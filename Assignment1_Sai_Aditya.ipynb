```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def buy_call(k,p,s):
  return np.maximum(0,s-k)-p

def buy_put(k,p,s):
  return np.maximum(0,k-s)-p

def sell_call(k,p,s):
  return p-np.maximum(0,s-k)

def sell_put(k,p,s):
  return p-np.maximum(0,k-s)
```

```python
k = float(input("Strike Price: "))
p = float(input("Premium: "))
range_width = min(100,k)
s = np.linspace(k-range_width,k+range_width,100)
```

```python
payoff = buy_call(k,p,s)
plt.plot(s,payoff,label="payoff")
plt.xlabel("Stock Price at expiry")
plt.ylabel("P&L - buying call option")
plt.legend()
```

As we are buying a call option, we are betting that the price will go up. Hence, up until the strike price, we are making a loss equal to the premium we spent, after which - we stay in loss until the stock price compensates for the premium spent, thereafter making a profit.

```python
payoff = sell_call(k,p,s)
plt.plot(s,payoff,label="payoff")
plt.xlabel("Stock Price at expiry")
plt.ylabel("P&L - selling call option")
plt.legend()
```

As we are selling a call option, we are betting that the price will go down. As a seller, the obvious, and the maximum profit we will be making is the premium given by the buyer. If the stock price increases, we stay in profit until the stock price is strike price + premium, and loss-making thereafter

```python
payoff = buy_put(k,p,s)
plt.plot(s,payoff,label="payoff")
plt.xlabel("Stock Price at expiry")
plt.ylabel("P&L - buying put option")
plt.legend()
```

When we buy a put option, we believe that the underlying asset's price will go down. Hence, we make profit until the underlying price dip is equal to the premium spent, after which we make a loss - and the maximum loss a buyer can make is the premium, forming a straight line at -20

```python
payoff = sell_put(k,p,s)
plt.plot(s,payoff,label="payoff")
plt.xlabel("Stock Price at expiry")
plt.ylabel("P&L - selling put option")
plt.legend()
```
